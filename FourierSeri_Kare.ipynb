```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,2*np.pi,5000,endpoint=False)
kare = signal.square(t)

plt.plot(t, kare)
plt.plot(t, 4/np.pi*(np.sin(t)+np.sin(3*t)/3))
```

```python
c = 4/np.pi
fourier = np.zeros_like(t)
for i in range(1,100000,2):
  fourier += np.sin(i*t)/i

fourier*=c

plt.plot(t, kare)
plt.plot(t, fourier)

```
